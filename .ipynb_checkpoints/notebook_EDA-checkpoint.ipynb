# Analyse Fond

## modification dataset, ajout distance big city

```python
#pip install geopy
```

```python
sf_coords = (37.77, -122.41) 
la_coords = (34.05, -118.24)

def distance_proche_big_city(row):
    city_coords = (row['latitude'], row['longitude'])
    distance_sf = geodesic(sf_coords, city_coords).kilometers
    distance_la = geodesic(la_coords, city_coords).kilometers
    return min(distance_sf, distance_la)

df['distance_nearest_city'] = df.apply(distance_proche_big_city, axis=1)

display(df)
```
